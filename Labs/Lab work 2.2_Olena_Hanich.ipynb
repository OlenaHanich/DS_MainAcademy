### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import statistics
import numpy
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
def minmax_foo():
    return min(list_with_random_numbers), max(list_with_random_numbers)
def mean_foo():
    return statistics.mean(list_with_random_numbers)
def std_foo():
    return numpy.std(list_with_random_numbers)
def median_foo():
    return statistics.median(list_with_random_numbers)
def count_foo():
    numbers_dict = {}
    for n in list_with_random_numbers:
        if n in numbers_dict:
            numbers_dict[n] += 1
        else: numbers_dict[n] = 1
    return numbers_dict
print('Мінімум дорівнює  ', minmax_foo()[0], 'Максимум дорівнює ', minmax_foo()[1])
print('Середнє значення дорівнює  ', mean_foo())
print('Медіана дорівнює ', median_foo())
print('Середнє квадратичне відхилення дорівнює ', std_foo())
print('Частото появи чисел в послідовності: ', count_foo())
```
